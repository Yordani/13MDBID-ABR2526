### Distribución de la variable objetivo

Es importante observar el balance de clases en la variable `y` (respuesta del cliente a la campaña).
Esto permite anticipar si habrá desbalance que afecte el modelo predictivo.


```python
sns.countplot(x="y", data=df)
plt.title("Distribución de la variable objetivo (suscripción al depósito)")
plt.xlabel("¿Suscribió un depósito a plazo?")
plt.ylabel("Cantidad de clientes")
plt.show()

df["y"].value_counts(normalize=True).mul(100).round(2)
```

### Distribución de variables categóricas

Las variables categóricas permiten entender mejor la composición del conjunto de clientes.

```python
categorical_cols = df.select_dtypes(include=["object"]).columns.drop("y")

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    order = df[col].value_counts().index
    sns.countplot(y=col, data=df, order=order)
    plt.title(f"Distribución de {col}")
    plt.xlabel("Cantidad")
    plt.ylabel(col)
    plt.show()
```

### Correlación entre variables numéricas

La correlación entre variables numéricas permite identificar relaciones lineales entre ellas y, consiguientemente, patrones que pueden ser útiles en modelos predictivos.

```python
num_df = df.select_dtypes(include=['float64', 'int64'])
corr = num_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlaciones')
plt.show()
```

### Distribución de variables numéricas

La distribución de variables numéricas permite observar cómo se comportan los datos en términos de frecuencia, dispersión y forma.
A través de histogramas con curva de densidad se pueden identificar sesgos, concentraciones, asimetrías y posibles valores atípicos que pueden influir en el modelado posterior.

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color='steelblue')
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()
    plt.close()
```

### Distribución de edad según aceptación del depósito (y)

Este gráfico permite analizar la relación entre la edad de los clientes y la variable objetivo y, que indica si aceptaron o no la oferta del depósito a plazo.
A través del boxplot se puede visualizar la mediana, el rango intercuartílico y la presencia de valores extremos, facilitando la comparación de la distribución de edades entre los grupos que respondieron “sí” y “no”.

```python
if 'y' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='y', y='age', data=df, palette='Set2')
    plt.title("Distribución de edad según aceptación del depósito (y)")
    plt.xlabel("Aceptó depósito (y)")
    plt.ylabel("Edad")
    plt.show()
```
